# 자료수집 - 데이터확보 

```python
!pip install yfinance
import yfinance as yf
import pandas as pd
start_date = '2000-01-01'
end_date = '2022-12-30'
data = yf.download(['USDKRW=X','JPYKRW=X'],start=start_date, end=end_date)
data.to_excel(f'{start_date}_{end_date}.xls')
data
```


# 자료수집 - 데이터 전처리

```python
USDKRW = []
JPYKRW = []
for a in data.columns:
  if a[1] == "USDKRW=X":
    USDKRW.append(a)
  elif a[1] == 'JPYKRW=X':
    JPYKRW.append(a)
data[USDKRW].to_excel(f'{start_date}_{end_date}_USDKRW.xls')
data[JPYKRW].to_excel(f'{start_date}_{end_date}_JPYKRW.xls')
```

# 통계분석&패턴파악- 데이터 분석

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,8))
plt.plot(data[USDKRW].index,data[USDKRW][('Adj Close', 'USDKRW=X')])
plt.title('Daily sales')
plt.xlabel('Date')
plt.ylabel('Sales')
```

```python
#데이터 시프트를 통해서 과거 20개의 데이터를 열값으로 만들고, 현재환율을 타겟값으로 변경
import datetime
year=str(datetime.datetime.now().year)
month=str(datetime.datetime.now().month)
day=str(datetime.datetime.now().day)
today = str("research_day:"+year+"/"+month+"/"+day)
new_data = data[[('Adj Close', 'USDKRW=X'),('Adj Close', 'JPYKRW=X')]]
new_data.columns = new_data.columns.droplevel(0)
for a in range(20):
  new_data[f'USDKRW=X-{str(a+1)}']=new_data['USDKRW=X'].shift(-1*(a+1))
new_data = new_data.dropna()
train = new_data[:-100].drop(columns="JPYKRW=X").iloc[:,1:]
test = new_data[:-100].drop(columns="JPYKRW=X")['USDKRW=X']
actual_train = new_data[-100:].drop(columns="JPYKRW=X").iloc[:,1:]
actual_test = new_data[-100:].drop(columns="JPYKRW=X")['USDKRW=X']
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from lightgbm import LGBMRegressor

DTR = LGBMRegressor()
DTR.fit(train,test)
pred = DTR.predict(actual_train)
plt.figure(figsize=(20,12))
plt.title(today)
plt.plot(test.index[-200:], test[-200:], 'o-')
plt.plot(actual_test.index, pred, 'o-', color='red', label='Predicted')
plt.plot(actual_test.index, actual_test, 'o-', color='black', alpha=0.2, label='Actual')
plt.legend(fontsize = 50)
plt.show()
```
